```python
import pandas as pd



# Función para eliminar las palabras de la lista en el texto
def eliminar_palabras(row):
    texto = row['texto']
    lista = row['hashtags']
    for palabra in lista:
        texto = texto.replace(palabra, '')
    return texto.strip()

# Aplicar la función a cada fila
df['texto_limpio']=df['texto'] = df.apply(eliminar_palabras, axis=1)
```

```python
df.head(3)
```

```python
import pandas as pd
from collections import Counter




# Aplanar la lista y contar la frecuencia de cada elemento
conteo_ht = Counter([elemento for sublist in df['hashtags'] for elemento in sublist])

# Crear un nuevo DataFrame a partir del conteo
df_frec_ht = pd.DataFrame.from_dict(conteo_ht, orient='index').reset_index()
df_frec_ht.columns = ['Elemento', 'Frecuencia']
```

```python
#Tabla de frecuencia de los Hashtags mas populares
df_frec_ht.sort_values('Frecuencia', ascending=False)
```

```python
def get_part_of_day(hour):
    if (hour >= 5) and (hour <= 11):
        return 'mañana'
    elif (hour > 11) and (hour <= 18):
        return 'tarde'
    else:
        return 'noche'

# Supongamos que 'df' es tu DataFrame y 'time' es tu columna de tiempo
df['part_of_day'] = df['fecha'].dt.hour.apply(get_part_of_day)
```

```python
import math

# Definimos los parámetros
N = 290606  # tamaño de la población
Z = 1.96  # valor Z para un 95% de confianza
p = 0.5  # proporción estimada del atributo en la población
e = 0.05  # margen de error tolerable

# Calculamos el tamaño de la muestra
n = (N * Z**2 * p * (1-p)) / (e**2 * (N-1) + Z**2 * p * (1-p))
n = math.ceil(n)  # redondeamos al entero más cercano

print(f'El tamaño de la muestra es {n}')
```
